```python
#a)Read the file to python IDE
import pandas as pd 
df=pd.read_csv('IBM_Details.csv')
 
```

```python
#b)Displaying the dataset
df
```

```python
 #c)Display the shape of dataset
df.shape
```

```python
#d)Display the datatypes of dataset
df.dtypes
```

```python
#e)Finding mean,median & mode of all the numeric data
print('\nMean')
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].mean())
print('\nMode')
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].mode())
print('\nMedian')
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].median())
```

```python
#f)Describe the entire dataset in terms of count,max,min,25%,50%,75%,std,var
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].count())
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].max())
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].min())
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].std())
print(df[['DailyRate','DistanceFromHome','WorkLifeBalance','YearsAtCompany','TotalWorkingYears']].var())
```
