```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('user_data.csv')
df
```

```python
x=df.iloc[:,[2,3]].values
y=df.iloc[:,4].values
 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(x,y, train_size= 2/3,test_size=1/3, random_state=0,shuffle=True) 

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#Transformation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(X_train) 
xtest = scaler.transform(X_test)
xtrain
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain, y_train)
```

```python
print("The intercept b0= ", classifier.intercept_)

print("The coefficient b1 = ", classifier.coef_[0][0])
print("The coefficient b2 = ", classifier.coef_[0][1])
```

```python
y_pred = classifier.predict(xtest)
y_pred
```

```python
#Need to explore on this--------
y_pred == y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
  
print ("Confusion Matrix : \n", cm)
```

```python
import seaborn as sns
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="BuPu")

#cmap="YlGnBu", cmap="Reds", cmap="BuPu"
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = xtest, y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1,stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1,stop = X_set[:, 1].max() + 1, step = 0.01))
  
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), alpha = 0.75, cmap = ListedColormap(('red', 'green')))
  
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
```

```python
from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
#Importing the dataset
iris=datasets.load_iris()
print(iris)
```

```python
X = iris.data[:,[2,3]]
print(X)
y=iris.target
print(y)
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1,stratify=y)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)

ppn=Perceptron(eta0=0.1,random_state=1)
ppn.fit(X_train_std,y_train)

y_pred=ppn.predict(X_test_std)
print(y_pred,',',y_test)
print('Accuracy: %.3f'% accuracy_score(y_test,y_pred))
print('Accuracy: %.3f'% ppn.score(X_test_std,y_test))
```
