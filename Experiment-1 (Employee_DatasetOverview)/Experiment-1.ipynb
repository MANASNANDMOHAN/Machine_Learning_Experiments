```python
#1)Import the employee dataset in Python IDE.
import pandas as pd
df=pd.read_csv('employee_dataset.csv')
df
```

```python
#2)Find the mean, median, mode, standard deviation, variance of any 5 attributes of the employee dataset.
print("Mean")
print(df[['Salary','MarriedID','GenderID','DeptID','EmpStatusID']].mean())
print("\nMedian")
print(df[['Salary','MarriedID','GenderID','DeptID','EmpStatusID']].median())
print("\nMode")
print(df[['Salary','MarriedID','GenderID','DeptID','EmpStatusID']].mode())
print("\nStandard Deviation")
print(df[['Salary','MarriedID','GenderID','DeptID','EmpStatusID']].std())
print("\nVariance")
print(df[['Salary','MarriedID','GenderID','DeptID','EmpStatusID']].var())
 
```

```python
#3) Plot the graph using Employee Dataset.
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,12)
df.plot(kind='density',subplots=True,layout=(5,5),sharex=False)
plt.savefig('Machine Learning.png')
plt.show()
 
```
